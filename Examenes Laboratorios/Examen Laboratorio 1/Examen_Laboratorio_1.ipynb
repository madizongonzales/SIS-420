Normalizacion de Caracteristicas

```python
def  featureNormalize(X):
    X_norm = X.copy()

    #Array de ceros con una longitud igual al n√∫mero de columnas de la matriz X
    mu = np.zeros(X.shape[1])
    sigma = np.zeros(X.shape[1])

    #Creamos el promedio de la matriz X
    mu = np.mean(X, axis = 0)
    #Desviacion estandar
    sigma = np.std(X, axis = 0)

    sigma[sigma == 0] = 1

    #Normalizacion de los datos
    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma
```

Funcion Sigmoide

```python
def sigmoid(z):
    z = np.array(z)
    # Calcula la sigmoide de una entrada z y convierte la intrada a un arreglo numpy
    g = np.zeros(z.shape)

    g = 1 / (1 + np.exp(-z))

    return g
```

Funcion del Costo

```python
def lrCostFunction(theta, X, y, lambda_):
    #Numero de ejemplos de entrenamiento igual al tamanio de las etiqudas de y
    m = y.size

    #Conviertir las etiquetas a valores enteros si son valores boleanos
    if y.dtype == bool:
        y = y.astype(int)

    #Almacenar valor de costo y gradiente
    J = 0
    grad = np.zeros(theta.shape)

    #Calcula la hipotesis
    h = sigmoid(X.dot(theta.T))

    #Para excluir el termino de sesgo en la regularizacion
    temp = theta
    temp[0] = 0

    J = (1 / m) * np.sum(-y.dot(np.log(h)) - (1 - y).dot(np.log(1 - h))) + (lambda_ / (2 * m)) * np.sum(np.square(temp))

    grad = (1 / m) * (h - y).dot(X)

    grad = grad + (lambda_ / m) * temp

    return J, grad
```

Clasificacion One-vs-All

```python
def oneVsAll(X, y, num_labels, lambda_):
    #Numero de ejemplos de entrenamiento
    m, n = X.shape

    #Inicializar la matriz de thetas con ceros, y con n+1 columnas (termino de sesgo)
    all_theta = np.zeros((num_labels, n + 1))

    #Agregar columna de '1' a la matriz  de caracteristicas X, pra calcular la hipotesis con el termino de sesgo
    X = np.concatenate([np.ones((m, 1)), X], axis=1)

    #Entrena un clasificador de regresion logistica para cada clase, utilizando un enforque de 'uno contra todos' devolviendo parametros optimos
    for c in np.arange(num_labels):
        initial_theta = np.zeros(n + 1)
        options = {'maxiter': 1000}
        res = optimize.minimize(lrCostFunction,
                                initial_theta,
                                (X, (y == c), lambda_),
                                jac=True,
                                method='CG',
                                options=options)

        all_theta[c] = res.x

    return all_theta
```

Prediccion One-vs-All

```python
def predictOneVsAll(all_theta, X):
    m = X.shape[0];
    num_labels = all_theta.shape[0]

    p = np.zeros(m)

    #Columna de '1' para el termino de sesgo, a la matriz de caracteristicas X
    X = np.concatenate([np.ones((m, 1)), X], axis=1)

    #Calcula las predicciones para cada ejemplo
    p = np.argmax(sigmoid(X.dot(all_theta.T)), axis = 1)

    return p
```
